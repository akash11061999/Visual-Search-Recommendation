```python
import glob
import json
import os
import cv2
```

```python
base_dir = "/home/akash/Downloads/Major_Project"
meta_dir = os.path.join(base_dir, "meta", "json")
structured_dir = os.path.join(base_dir, "structured_images")
query_files = glob.glob(meta_dir + "/*_pairs_*.json")
```

```python
count=0
failed=[]
for path in query_files:
    vertical = path.split("_")[-1].split(".")[0]
    wtbi_crop_dir = os.path.join(structured_dir, "wtbi_"+vertical+"_query_crop")
    if not os.path.exists(wtbi_crop_dir):
        os.mkdir(wtbi_crop_dir)
    query_dir = os.path.join(structured_dir, vertical+"_query")
    print("Processing path %s"%(path))
    with open(path) as jsonFile:
        pairs = json.load(jsonFile)
    #print(pairs)
    for pair in pairs:
        query_id = str(pair["photo"]).zfill(9)
        bbox = pair["bbox"]
        flag = False
        try:
            for extension in [".jpeg", ".png", ".jpg"]:
                query_path = os.path.join(query_dir, query_id + extension)
                if os.path.exists(query_path):
                    img = cv2.imread(query_path, cv2.IMREAD_COLOR)
                    x, w, y, h = bbox["left"], bbox["width"], bbox["top"], bbox["height"]
                    crop_img = img[y:y+h, x:x+w]
                    cv2.imwrite(os.path.join(wtbi_crop_dir, query_id+extension), crop_img)
                    flag = True
                    count+=1
                    break
            if not flag:
                failed.append(query_id)
        except:
            failed.append(query_id)
            continue
        
#         try:
#             if os.path.exists(query_path_1):
#                 img = cv2.imread(query_path_1, cv2.IMREAD_COLOR)
#                 x, w, y, h = bbox["left"], bbox["width"], bbox["top"], bbox["height"]
#                 crop_img = img[y:y+h, x:x+w]
#                 cv2.imwrite(os.path.join(wtbi_crop_dir, query_id+".jpg"), crop_img)
#             elif os.path.exists(query_path_2):
#                 img = cv2.imread(query_path_2, cv2.IMREAD_COLOR)
#                 x, w, y, h = bbox["left"], bbox["width"], bbox["top"], bbox["height"]
#                 crop_img = img[y:y+h, x:x+w]
#                 cv2.imwrite(os.path.join(wtbi_crop_dir, query_id+".jpeg"), crop_img)
#             elif os.path.exists(query_path_3):
#                 img = cv2.imread(query_path_3, cv2.IMREAD_COLOR)
#                 x, w, y, h = bbox["left"], bbox["width"], bbox["top"], bbox["height"]
#                 crop_img = img[y:y+h, x:x+w]
#                 cv2.imwrite(os.path.join(wtbi_crop_dir, query_id+".png"), crop_img)
#             else:
#                 failed.append(query_id)
#         except:
#             failed.append(query_id)
print(count, len(failed))
```
